# Create Schema

Note - We will create a schema for all the tables, but we will not establish any relationships yet. The data needs to be loaded into the SQL server first, and after cleaning it for consistency, we will then establish the necessary relationships.

```python
#1. geolocation
c_geolocation = """
CREATE TABLE IF NOT EXISTS geolocation (
geolocation_zip_code_prefix INT,
geolocation_lat DOUBLE PRECISION,
geolocation_lng DOUBLE PRECISION,
geolocation_city VARCHAR,
geolocation_state CHAR(2)
);
"""

with engine.begin() as conn:
    conn.execute(text(c_geolocation))
```

```python
#2. customer 
c_customer = """
CREATE TABLE customers (
customer_id VARCHAR(50) PRIMARY KEY,
customer_unique_id VARCHAR(50),
customer_zip_code_prefix INT,
customer_city VARCHAR,
customer_state CHAR(2));
"""

with engine.begin() as conn:
    conn.execute(text(c_customer))
```

```python
#3. orders
c_orders = """
CREATE TABLE IF NOT EXISTS orders (
order_id VARCHAR(50) PRIMARY KEY,
customer_id VARCHAR(50),
order_status VARCHAR(20),
order_purchase_timestamp TIMESTAMP,
order_approved_at TIMESTAMP,
order_delivered_carrier_date TIMESTAMP,
order_delivered_customer_date TIMESTAMP,
order_estimated_delivery_date TIMESTAMP);
"""

with engine.begin() as conn:
    conn.execute(text(c_orders))
```

```python
#4. order_payments
c_order_payments = """
CREATE TABLE IF NOT EXISTS order_payments (
order_id VARCHAR(50),
payment_sequential INT,
payment_type VARCHAR(30),
payment_installments INT,
payment_value NUMERIC(10,2));
"""

with engine.begin() as conn:
    conn.execute(text(c_order_payments))
```

```python
#5. order_reviews
c_order_reviews = """
CREATE TABLE IF NOT EXISTS order_reviews (
review_id VARCHAR(50) PRIMARY KEY,
order_id VARCHAR(50),
review_score INT,
review_comment_title TEXT,
review_comment_message TEXT,
review_creation_date TIMESTAMP,
review_answer_timestamp TIMESTAMP);
"""

with engine.begin() as c:
    c.execute(text(c_order_reviews))
```

```python
#6. sellers
c_sellers = """
CREATE TABLE IF NOT EXISTS sellers (
seller_id VARCHAR(50) PRIMARY KEY,
seller_zip_code_prefix INT,
seller_city VARCHAR,
seller_state CHAR(2));
"""

with engine.begin() as c:
    c.execute(text(c_sellers))
```

```python
#7. product_category
c_product_category = """
CREATE TABLE IF NOT EXISTS product_category (
ï»¿product_category_name VARCHAR PRIMARY KEY,
product_category_name_english VARCHAR
);
"""

with engine.begin() as c:
    c.execute(text(c_product_category))
```

```python
#8. products
c_products = """
CREATE TABLE IF NOT EXISTS products (
product_id VARCHAR(50) PRIMARY KEY,
product_category_name VARCHAR,
product_name_lenght INT,
product_description_lenght INT,
product_photos_qty INT,
product_weight_g INT,
product_length_cm INT,
product_height_cm INT,
product_width_cm INT
);
"""

with engine.begin() as c:
    c.execute(text(c_products))
```

```python
#9. order_items
c_order_items = """
CREATE TABLE IF NOT EXISTS order_items (
order_id VARCHAR(50),
order_item_id INT,
product_id VARCHAR(50),
seller_id VARCHAR(50),
shipping_limit_date TIMESTAMP,
price NUMERIC(10,2),
freight_value NUMERIC(10,2),
PRIMARY KEY(order_id,order_item_id)
);
"""

with engine.begin() as c:
    c.execute(text(c_order_items))
```
